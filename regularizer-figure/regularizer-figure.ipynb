```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 
matplotlib.rc('font', family='serif')
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True
from palettable.colorbrewer.qualitative import *
```

```python
eps = 1e-6
x = np.linspace(-1.5, 1.5, 1001)
y1 = np.abs(x)
y2 = np.log(1+np.abs(x)/eps) / np.log(1+1/eps)
# y3 = np.ones_like(x)
# y3[500] = 0
```

```python
plt.plot(x, y1, lw=2, c=Paired_6.mpl_colors[1], label="$\ell_1$-norm regularizer")
plt.plot(x, y2, lw=2, c=Paired_6.mpl_colors[-1], label="log-norm regularizer")
# plt.plot(x, y3, label="$\ell_0$ norm", c='k')
plt.legend(fontsize=14)
plt.savefig('regularizer.pdf')
plt.show()
```
