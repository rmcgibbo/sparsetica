```python
from fipy import *
import numpy as np
%matplotlib inline
import sympy
import time
from matplotlib.mlab import griddata
from sympy.utilities.lambdify import lambdify
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
from palettable.colorbrewer.diverging import *
from matplotlib import rc
rc('font',**{'family':'serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

$$
\frac{dp(x,t)}{dt} = - \frac{d}{dx} (\nabla U(x) p(x,t)) + \Delta (D p(x,t))
$$

$$
\frac{dp(x,t)}{dt} = \frac{d}{dx} (F(x), p(x,t)) + \frac{d^2}{dx^2} (D p(x,t))
$$

$$ U(x,y) = [1 - 0.5 \tanh(y - x)](x + y - 5)^2 + 0.2[((y - x)^2 - 9)^2 + 3(y - x)] +15e^{-(x-2.5)^2-(y-2.5)^2}
- 20e^{-(x-4)^2-(y-4)^2}
$$

```python
def U(x,y):
    if isinstance(x, np.ndarray):
        tanh, exp = np.tanh, np.exp
    else:
        tanh, exp = sympy.tanh, sympy.exp

    term1 = (1 - 0.5*tanh(y - x))*(x + y - 5)**2
    term2 = 0.2*(((y - x)**2 - 9)**2 + 3*(y - x))
    term3 = 15*exp(-(x-2.5)**2 - (y-2.5)**2)
    term4 = -20*exp(-(x-4)**2 - (y-4)**2)
    return term1 + term2 + term3 + term4


def build_grad():
    x, y = sympy.var('x y')
    E = U(x,y)
    grad = lambdify((x, y), (sympy.diff(E, x)/5., sympy.diff(E, y)/5.))
    return grad

grad = build_grad()
```

```python
def plot_potential(linestyles='-', show_labels=True):
    X,Y = np.meshgrid(np.linspace(-4,8,100), np.linspace(-2,8,100))
    Z = U(X, Y)

    CS = plt.contour(X, Y, Z, levels=[0,5,10,15,20,25,30], colors='k', linestyles=linestyles)
    if show_labels:
        plt.clabel(CS, [5,10], inline=1, fontsize=12, fmt='%1.f')
    plt.xlim(-3.5,7)
    plt.grid(False)

plt.figure(figsize=(4,4))
plot_potential()
plt.gca().set_xlabel(r'$x$', fontsize=20)
plt.gca().set_ylabel(r'$y$', fontsize=20)
for tick in plt.gca().xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in plt.gca().yaxis.get_major_ticks():
    tick.label.set_fontsize(14)


plt.annotate('', (1,4), xytext=(4.2, 0.9),
    arrowprops=dict(arrow_defaults, lw=3, color='#4dac26', arrowstyle='->', connectionstyle='arc3,rad=.7'))
plt.annotate('', (0.8,3.8), xytext=(4,1),
    arrowprops=dict(arrow_defaults, lw=3, color='#d01c8b', arrowstyle='->', connectionstyle='arc3,rad=-.7'))

plt.tight_layout()
plt.savefig('2d-potential.pdf')
```

```python
from fipy.meshes.uniformGrid2D import UniformGrid2D

n_cells = 50
x_limits = (-4.,8.)
y_limits = (-2.,8.)

mesh = UniformGrid2D(
    dx=(x_limits[1]-x_limits[0])/n_cells,
    dy=(y_limits[1]-y_limits[0])/n_cells,
    nx=n_cells,
    ny=n_cells,
    origin=((x_limits[0],), (y_limits[0],)))
```

```python
gradOnMesh = np.array([grad(x, y) for (x, y) in zip(mesh.cellCenters[0], mesh.cellCenters[1])])
gradVariable = CellVariable(
    name='grad', mesh=mesh,
    value=gradOnMesh.T)
eq = TransientTerm() == DiffusionTerm(coeff=1) + ConvectionTerm(coeff=gradVariable)
```

```python
energies = U(np.asarray(mesh.cellCenters[0]), np.asarray(mesh.cellCenters[1]))
weights = np.exp(-energies/5)
Z = np.sum(weights)
mu2 = weights / Z
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=5, include_bias=False)
```

```python
cellx = np.asarray(mesh.cellCenters[0])
celly = np.asarray(mesh.cellCenters[1])
features = pf.fit_transform(np.vstack([cellx-np.mean(cellx), celly-np.mean(celly)]).T)
```

```python
C = np.zeros((features.shape[1], features.shape[1]))
S = np.zeros((features.shape[1], features.shape[1]))

for i in range(features.shape[1]):
    for j in range(features.shape[1]):
        c, s = correlation(features[:, i], features[:, j])
        C[i,j] = c
        S[i,j] = s
```

```python
tica_eigs, tica_vecs = scipy.linalg.eig(C,S)
```

```python
print(np.argsort(tica_eigs))
print(np.max(tica_eigs))
print(tica_eigs[1])
tica_vecs[:, 1]
```

```python
a,b = correlation(q,q)
print(a/b)
a, b = correlation(features.dot(tica_vecs[:, 1]), features.dot(tica_vecs[:, 1]))
print(a/b)
```
